```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset.head()
```

```python
dataset = dataset.drop(['User ID', 'Gender'], axis=1)

print(dataset.head())
```

```python
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=dataset)
plt.show()
```

```python
x = dataset.iloc[:, :-1]
y = ["Purchased"]
```

```python
sc = StandardScaler()
sc.fit(x)
x = pd.DataFrame(sc.transform(x), columns=x.columns)
x
```

```python
# x = features (all rows, all columns EXCEPT the last one)
x = dataset.iloc[:, :-1].values

# y = target (all rows, ONLY the last column)
y = dataset.iloc[:, -1].values

# Now, let's check the shapes (this is for your reference)
print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)*100
```

```python
knn.score(x_train, y_train)*100
```

```python
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    print(f"Accuracy for k={i}: {knn.score(x_test, y_test)*100}")
```

```python
knn.predict([[-1.781797, -1.490046]])
```

```python
plot_decision_regions(x_test, y_test, clf=knn, legend=2)
plt.show()
```

## K Nearest Neighbor Regression 

```python
dataset = pd.read_csv('multiple_linear_regression_dataset.csv')
dataset.head()
```

```python
x = dataset.drop(columns = "income")
y = dataset["income"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)*100
```
