```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.svm import SVR
```

```python
dataset = pd.read_csv("polynomial classification.csv")
dataset.head()
```

```python
sns.scatterplot(x = "x1", y = "x2", data = dataset, hue = "label")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["label"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sv = SVC(kernel = "poly")
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=sv, legend=2)
plt.show()
```

## Regression

```python
dataset1 = pd.read_csv("placement.csv")
dataset1.head()
```

```python
sns.scatterplot(x = "cgpa", y = "package", data = dataset1)
plt.show()
```

```python
x = dataset1[["cgpa"]]
y = dataset1[["package"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
'linear', 'poly', 'rbf', 'sigmoid'
```

```python
sv = SVR(kernel = "linear")
sv.fit(x_train, y_train)
```

```python
sv.score(x_test, y_test)*100
```

```python
sv.score(x_train, y_train)*100
```

```python
sns.scatterplot(x = "cgpa", y = "package", data = dataset1)
plt.plot(dataset1["cgpa"], sv.predict(x), color = "red")
plt.show()
```
