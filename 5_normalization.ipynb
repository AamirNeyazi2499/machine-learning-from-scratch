```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('loan.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.displot(dataset['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_min'] = ms.transform(dataset[['CoapplicantIncome']])
dataset.head()
```

```python
''' plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Before Normalization')
sns.displot(dataset['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After Normalization')
sns.displot(dataset['CoapplicantIncome_min']) '''

plt.figure(figsize=(12, 5))

ax1 = plt.subplot(1, 2, 1)
ax1.set_title('Before Normalization')
sns.histplot(data=dataset, x='CoapplicantIncome', ax=ax1)

ax2 = plt.subplot(1, 2, 2)
ax2.set_title('After Normalization')
sns.histplot(data=dataset, x='CoapplicantIncome_min', ax=ax2)

plt.tight_layout()
plt.show()
```
